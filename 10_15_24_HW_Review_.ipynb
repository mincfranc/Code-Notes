````python
# Define the columns
B = ['Street', 'AnotherColumn']  # Example of multiple columns in B
A = ['Alley']

# Count values for A and B
counts_A = categorical_cols[A].value_counts(dropna=True).reset_index()
counts_B = categorical_cols[B].value_counts(dropna=True).reset_index()

# Merge the dataframes on multiple columns in B
merged_df = pd.merge(
    counts_A,
    counts_B,
    on=B,
    suffixes=('_A', '_B')
)

# Calculate marginal probabilities
merged_df['marg_prob'] = merged_df['count_A'] / merged_df['count_B'] * 100

# Display the resulting dataframe
merged_df
````
